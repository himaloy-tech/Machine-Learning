<h2>Data Wrangling</h2>

```python
import pandas as pd
filename = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv'
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(filename, names = headers)
df.head(10)
```

```python
import numpy as np
df.replace("?", np.nan, inplace=True)
df.head(10)
```

```python
missing_data = df.isnull()
missing_data.head(10)
```

```python
for column in missing_data.columns.tolist():
    print(column)
    print(missing_data[column])
    print(" ")
```

```python
for column in missing_data.columns.tolist():
    print(column)
    print(missing_data[column].value_counts())#True = Blank False = Not Blank
    print(" ")
```

```python
avg_normaloss = df["normalized-losses"].astype("float").mean(axis=0)
print(avg_normaloss)
```

```python
df["normalized-losses"].replace(np.nan, avg_normaloss, inplace=True)
df.head(10)
```

```python
mode_noofdoors = df["num-of-doors"].value_counts().idxmax()
print(mode_noofdoors)
```

```python
df["num-of-doors"].replace(np.nan, mode_noofdoors, inplace=True)
df.head(10)
```

```python
df.dropna(subset = ["price"], axis=0, inplace=True)
df.head(10)#delete row
```

```python
#Reset index after dropna
df.reset_index(drop = True, inplace=True)
df.head(10)
```

<h2>Changing Data Format</h2>

```python
df.dtypes
```

```python
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df.dtypes
```

```python
df[["bore", "stroke", "price", "peak-rpm"]] = df[["bore", "stroke", "price", "peak-rpm"]].astype("float")
df.dtypes
```

<h2>Data Strandardization</h2>

```python
df["city-mpg"] = 235/df["city-mpg"]
df.head(10)
```

```python
df.rename(columns={"city-mpg":"city-l/100km"}, inplace=True)
df.head(10)
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns = {"highway-mpg":"highway-l/100km"}, inplace=True)
df.head(10)
```

<h2>Data Normalization</h2>

```python
df["height"]
```

```python
df.columns
```

```python
df[["height", "width", "length"]].head(10)
```

```python
df["height"] = df["height"]/df["height"].max()
df["width"] = df["width"]/df["width"].max()
df["length"] = df["length"]/df["length"].max()
df[["height", "width", "length"]].head(10)
```

```python
horsepower_mean = df["horsepower"].astype('float').mean(axis=0)
df["horsepower"].replace(np.nan, horsepower_mean, inplace=True)
df["horsepower"] = df["horsepower"].astype('int')
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"]) # Histogram
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
import numpy as np
bins = np.linspace(df["horsepower"].min(), df["horsepower"].max(), num=4)
groupnames = ['low', 'med', 'high']
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels=groupnames, include_lowest = True)
df[["horsepower", "horsepower-binned"]].head(10)
```

```python
print(df["horsepower-binned"].value_counts())
```

```python
plt.pyplot.hist(df["horsepower-binned"])
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```
