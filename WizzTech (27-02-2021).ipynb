```python
drive_wheels_count=df["drive-wheels"].value_counts().to_frame()
```

```python
drive_wheels_count.rename(columns={"drive-wheels":"value-counts"}, inplace=True)
drive_wheels_count
```

```python
drive_wheels_count.index.name = 'drive-wheels'
drive_wheels_count
```

```python
style = df["body-style"].value_counts().to_frame()
style.rename(columns={"body-style":"count"}, inplace=True)
```

```python
style.index.name = 'style'
```

```python
style
```

```python
#Basics of grouping
```

```python
df['body-style'].unique()
```

```python
df_group_one = df[["body-style", "price"]]
df_groupone = df_group_one.groupby(['body-style'], as_index=False).mean()
df_groupone
```

```python
df_gptest = df[["drive-wheels", "body-style", "price"]]
grouped_test1 = df_gptest.groupby(["drive-wheels", "body-style"], as_index=False).mean()
grouped_test1
```

```python
#Pivot Table
grouped_pivot = grouped_test1.pivot(index  = 'drive-wheels', columns = 'body-style')
grouped_pivot
```

```python
#Fills NaN with a value
grouped_pivot = grouped_pivot.fillna(-1)
grouped_pivot
```

```python
name = df[["drive-wheels", "price"]]
result = name.groupby(["drive-wheels"], as_index = False).mean()
result
```

```python
#Heat Map
plt.pcolor(grouped_pivot, cmap="RdBu")
plt.colorbar()
plt.show()
```

```python
fig,ax=plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')
#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index
#Move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0])+0.5, minor=False)
#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)
#rotate label if too long
plt.xticks(rotation=90)
fig.colorbar(im)
plt.show()
```
