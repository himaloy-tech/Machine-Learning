```python
#Value Counts
```

```python
df["drive-wheels"].value_counts()
```

```python
drive_wheels_count=df["drive-wheels"].value_counts().to_frame()
```

```python
drive_wheels_count.rename(columns={"drive-wheels":"value-counts"}, inplace=True)
drive_wheels_count
```

```python
drive_wheels_count.index.name = 'drive-wheels'
drive_wheels_count
```

```python
style = df["body-style"].value_counts().to_frame()
style.rename(columns={"body-style":"count"}, inplace=True)
```

```python
style.index.name = 'style'
```

```python
style
```

```python
df_group_one = df[["body-style", "price"]]
df_groupone = df_group_one.groupby(['body-style'], as_index=False).mean()
df_groupone
```
